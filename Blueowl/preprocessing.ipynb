```python
import os
import numpy as np 
from numpy.random import RandomState
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
```

```python
df1 = pd.read_csv('Data/AW_AveMonthSpend.csv', delimiter=',', nrows = None)
df1.dataframeName = 'AW_AveMonthSpend.csv'
nRow, nCol = df1.shape
print(f'There are {nRow} rows and {nCol} columns')
df1.head(5)
```

```python
df2 = pd.read_csv('Data/AW_BikeBuyer.csv', delimiter=',', nrows = None)
df2.dataframeName = 'AW_BikeBuyer.csv'
nRow, nCol = df2.shape
print(f'There are {nRow} rows and {nCol} columns')
df2.head(5)
```

```python
df3 = pd.read_csv('Data/AdvWorksCusts.csv', delimiter=',', nrows = None)
df3.dataframeName = 'AdvWorksCusts.csv'
nRow, nCol = df3.shape
print(f'There are {nRow} rows and {nCol} columns')
df3.head(5)
```

```python
# df4 = pd.read_csv('Data/AW_test.csv', delimiter=',', nrows = None)
# df4.dataframeName = 'AW_test.csv'
# nRow, nCol = df4.shape
# print(f'There are {nRow} rows and {nCol} columns')
# df4.head(5)
```

```python
data = pd.concat([df3, df1['AveMonthSpend'], df2['BikeBuyer']], axis=1, join='inner').sort_index()
nRow, nCol = data.shape
print(f'There are {nRow} rows and {nCol} columns')
data.head()
```

```python
data.isnull().sum()
```

```python
data = data.drop(data.columns[[1, 3, 5, 7]], axis=1)
data.head(5)
```

```python
data['CustomerID'] = data.index
data.head()
```

```python
data['split'] = np.random.randn(data.shape[0], 1)
msk = np.random.rand(len(data)) <= 0.7
train = data[msk]
test = data[~msk]

train = train.drop(columns='split')
print(len(train), len(test))
```

```python
test2 = test.drop(columns='BikeBuyer')
test2 = test2.drop(columns='split')
test2.head()
```

```python
answer_key = test['BikeBuyer']
answer_key = pd.DataFrame(answer_key)
answer_key['CustomerID'] = answer_key.index
answer_key = answer_key[['CustomerID', 'BikeBuyer']]
answer_key
```

```python
sample_submission = test['BikeBuyer']
sample_submission = sample_submission.replace(0)
sample_submission = pd.DataFrame(sample_submission)
sample_submission['CustomerID'] = sample_submission.index
sample_submission = sample_submission[['CustomerID', 'BikeBuyer']]
sample_submission
```

```python
train.to_csv('final/train.csv', index=False)
test2.to_csv('final/test.csv', index=False)
answer_key.to_csv('final/solution.csv', index=False)
sample_submission.to_csv('final/sample_submission.csv', index=False)
```

### Label encoding

```python
train_data = pd.read_csv('final/train.csv', delimiter=',')
nRow, nCol = train_data.shape
print(f'There are {nRow} rows and {nCol} columns')
train_data.head(5)
```

```python
labelencoder = LabelEncoder()
for col in train_data.columns:
    train_data[col] = labelencoder.fit_transform(train_data[col])
train_data.head()
```

```python
test_data = pd.read_csv('final/test.csv', delimiter=',')
nRow, nCol = test_data.shape
print(f'There are {nRow} rows and {nCol} columns')
test_data.head(5)
```

```python
labelencoder = LabelEncoder()
for col in test_data.columns:
    test_data[col] = labelencoder.fit_transform(test_data[col])
test_data.head()
```

### Train & test encoded data

```python
train_data.to_csv('train_encode.csv', index=False)
test_data.to_csv('test_encode.csv', index=False)
```
